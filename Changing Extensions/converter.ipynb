# Sample Scripts to convert Image file extensions

```python
from PIL import Image
import cv2
  
#open image in png format
img_png = Image.open('Gallery\hard_hat_workers1340.png')
  
#The image object is used to save the image in jpg format
img_png.save('Gallery\hard_hat_workers1340.jpg')
```

```python
img_jpg = Image.open('Gallery\hard_hat_workers1340.jpg')
```

```python
print(type(img_png))
print(type(img_jpg))
```

```python
img_png == img_jpg
for i, j in zip(img_png,img_jpg):
    print(i==j)
```

```python
#importing required packages and library
import cv2
  
# Loading .png image
png_img = cv2.imread('Gallery\hard_hat_workers1341.png')
  
# converting to jpg file
# saving the jpg file
cv2.imwrite('Gallery\hard_hat_workers1341.jpg', png_img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```

```python
img_png = cv2.imread("Gallery\hard_hat_workers1341.png")
img_jpg = cv2.imread("Gallery\hard_hat_workers1341.jpg")
```

```python
print(img_png.shape)
print(img_jpg.shape)
```

```python
img_png == img_jpg
for i, j in zip(img_png,img_jpg):
    print(i==j)
```

From above analysis it is clear that the first method (extension conversion using PIL) has better performance. since, it causes less damage or changes to the image
